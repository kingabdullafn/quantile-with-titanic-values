```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Titanic Dataset.csv")

df.head(10)
```

```python
df.isnull().sum()
```

```python
age_q1 = np.quantile(df['Age'], 0.25)
age_q2 = np.quantile(df['Age'], 0.50)
age_q3 = np.quantile(df['Age'], 0.75)
```

```python
print("Age Quartiles ")
print("Q1 -", age_q1)
print("Q2 -", age_q2)
print("Q3 -", age_q3)
```

```python
IQR_age = age_q3 - age_q1
print("interquartile range :", IQR_age)
```

```python
plt.hist(df['Age'])
plt.ylabel("Count of passengers")
plt.xlabel("Age")
```

```python
fare_q1 = np.quantile(df['Fare'], 0.25)
fare_q2 = np.quantile(df['Fare'], 0.50)
fare_q3 = np.quantile(df['Fare'], 0.75)
```

```python
print("Fare Quartiles ")
print("Q1 -", fare_q1)
print("Q2 -", fare_q2)
print("Q3 -", fare_q3)
```

```python
IQR_fare = fare_q3 - fare_q1
print("interquartile range :", IQR_fare)
```

```python
bins =np.arange(0,250,20)
plt.hist(df['Fare'], bins=np.arange(1,250,20))
plt.ylabel("Count of passengers")
plt.xlabel("Fare")
plt.xticks(bins)
```

```python
plt.boxplot(df['Age'])
plt.title("Age distribution")
plt.show()
```

```python
plt.boxplot(df['Pclass'])
plt.title("Passenger Class distribution")
plt.show()
```
